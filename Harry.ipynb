```python
io = open('Harry Potter.txt', 'r')
text = io.read()
print(text)
io.close()
```

```python
#Clean up the text

import string

#Remove punctuation 
text = text.translate(str.maketrans('', '', string.punctuation)).lower().split()
```

```python
#I can improve this by destemming and removing plurals

import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

def cleanup(lis):
    wnl = WordNetLemmatizer()
    lis = [wnl.lemmatize(x) for x in lis]
    return lis
    
text = cleanup(text)
```

```python
dic = {k:text.count(k) for k in set(text)}
print(dic)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(dic, orient = 'index')
df.index.name = 'Word'
```

```python
df.rename({0: "Occurances"}, axis = 1, inplace = True)
df.sort_values(by = 'Word', inplace = True)
print(df)
```
